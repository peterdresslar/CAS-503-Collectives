# Identifying scaling laws

In lecture, we saw how scaling laws can often be used to relate quantities in collectives to one another.  A common basic type of scaling takes the form of a "self-similar" power law:
$$
A = k B^\beta,
$$
where $A$ and $B$ are two quantities of interest, $k$ is a constant factor, and $\beta$ is the exponent of the power law.

In this exercise, we'll look for scaling laws within some real-world ecological data from measurements of trees.

The function `plt.scatter` will take two columns of our dataset and plot points in a 2D scatter plot:

```python
# Hello, it's Peter (the other interlocutor) here.... adding this utility function to apply to both the base notebook and my responses.

def plt_util(x, y, tick_interval, pad=0.5, ax=None, start_at_zero=True):
    if ax is None:
        ax = plt.gca()

    # so that we can visually understand the absolute relationship, let's make sure our axes have the same range
    # if we want to be picky about it (and not just use 0,5) we can find the mins and maxes of *both* axes
    axmin = 0 if start_at_zero else min(numpy.min(x), numpy.min(y))
    axmax = max(numpy.max(x), numpy.max(y)) + pad
    ax.set_xlim(axmin, axmax)
    ax.set_ylim(axmin, axmax)

    ax.set_box_aspect(1)  # to be very particular about it
    # ensure we have tick marks for every integer using np.arange (and suppress the text output)
    _ = ax.set_xticks(numpy.arange(axmin, axmax, tick_interval))
    _ = ax.set_yticks(numpy.arange(axmin, axmax, tick_interval))

    return ax
```

```python
# load data
data = pandas.read_csv(Path('../data/McPhersonEtAl2020/TS3_Raw_tree_data.csv')) # caution: path is updated for my local fork
dataSubset = data[ (data['CommonName']=='Japanese zelkova') ]

# make scatter plot
plt.scatter(dataSubset['DBH (cm)'],dataSubset['TreeHt (m)'])
plt.xlabel('diameter (cm)')
plt.ylabel('height (m)')

_ax = plt_util(dataSubset['DBH (cm)'], dataSubset['TreeHt (m)'], 20) # underscore suppresses output in jupyter

```

This looks quite linear, indicating that tree height does scale (at least roughly) as a power law of diameter.

Recall that the slope of a line on a log-log plot indicates the power of the power law: taking the logarithm of both sides of the equation at the beginning of the notebook, we have
$$
\log A = \log k + \beta \log B,
$$
so the exponent $\beta$ in the original equation is the slope of a linear relationship between $\log A$ and $\log B$.

So we will now fit a line to our logged data, making use of a polynomial fitting function from `numpy` (a polynomial of degree 1 is a line):

```python
polyfit = numpy.polynomial.Polynomial.fit(numpy.log(dataSubset['DBH (cm)']),
                                          numpy.log(dataSubset['TreeHt (m)']),1).convert()
polyfit
```

This output means that the logarithm of tree height is approximately 0.39 plus 0.57 times the logarithm of DBH.

⚛️ **Bonus question (for nothing but bragging rights):** An important quantity for ecologists is the bio-mass, basically the total mass of the wood in our tree example (which is related, for example, to the total amount of carbon captured by a forest).  The bio-mass of a tree is harder to measure than, say, its diameter.  Use your scaling law and any needed basic assumptions to predict how a Japanese zelkova's biomass scales as a function of its diameter.  This relates our scaling analysis with ecologically relevant collective effects of the cells that combine to form a tree.

✴️ **Answer:** Hmmmm.... an interesting question.

Mass is going to be proportionate to the density of the tree's material and its absolute volume. I think wetness of the wood would totally skew mass, but I presume that there is some standard fix biologists have to factor out the plant moisture. So, density: wood/leaf mass per cubic meter (which I would guess would be pretty standardized in mature trees) times total volume.

We know fairly well how tall a Japanese zelkova tree will be for a given trunk size, but how to get a volume for that? Just as a guess, I could imagine that a conic pyramid would be a *reasonable* proxy: I am torn between that or a cylinder, but I suspect that leaf mass is seriously tied to water (in other words, I suspect that leaf *biomass* is pretty low.) So, let's try a conic pyramid, with of course the measured trunk size being assumed to be the base of the cone (rather than higher off the ground. It's a bit of a hack but nothing too unreasonable.)

The formula for a cone's volume is $V = \frac{1}{3} \pi r^2 h$.

So... we have a bit of a task here to come up with the formula for the volume of a tree as a function of its diameter. Note that we'll need to be cautious about radius versus diameter, here. 

Again let's just call `DBH` and `height` $x$ and $y$ for clarity:

$$
\begin{aligned}
V &= \frac{1}{3} \cdot \pi \cdot \frac{x}{2}^2 \cdot y \\
\end{aligned}
$$

we also know that
$$
\begin{aligned}
\ln(y) &= 0.39096 + 0.57336 \cdot \ln(x) \\
\end{aligned}
$$

from which we can isolate y...

$$
\begin{aligned}
y = e^{0.39096 + 0.57336 \ln(x)} \\
y = e^{0.39096} + e^{0.57336 \ln(x)} \\
y = 1.478 \cdot x^{0.57336} \\
\end{aligned}
$$

And thus, back to our Volume formula, we can now plug in y in terms of x...

$$
\begin{aligned}
V &= \frac{1}{3} \cdot \pi \cdot \frac{x}{2}^2 \cdot 1.478 \cdot x^{0.57336} \\
\end{aligned}
$$

We have a bit of a tricky conversion here owing to our radius converter. A trip to Khan Academy reminds me that we can apply the exponent to both numerator and denominator. And so $$\left(\frac{x}{2}\right)^2 = \frac{1}{4} \cdot x^2$$

This leaves us with a nice clean path to:

$$
\begin{aligned}
V &= 0.38794 \cdot x^{2.57336} \\
\end{aligned}
$$

That *looks* like an acceptable scaling equation, but it's important to remember that we are looking for mass. If, indeed, mass relates directly to volume (I don't see how it would not), then we would have another scalar that would combine with our 0.38794 scalar, but I do not think that these affect scalars alter the core power function itself. We should be safe to guess that the Mass of our tree scales at *about* the rate:

$$
\begin{aligned}
M \sim DBH^{2.57336} \\
\end{aligned}
$$
